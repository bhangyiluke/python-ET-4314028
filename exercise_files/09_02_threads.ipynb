```python
import threading
import time
```

## Threads

```python
def longSquare(num):
    time.sleep(1)
    return num**2

[longSquare(n) for n in range(0, 5)]
```

```python
t1 = threading.Thread(target=longSquare, args=(1,))
t2 = threading.Thread(target=longSquare, args=(2,))
t3 = threading.Thread(target=longSquare, args=(3,))
t4 = threading.Thread(target=longSquare, args=(4,))
t5 = threading.Thread(target=longSquare, args=(5,))

t1.start()
t2.start()
t3.start()
t4.start()
t5.start()

t1.join()
t2.join()
t3.join()
t4.join()
t5.join()
```

```python


def longSquare(num, results):
    time.sleep(1)
    results[num] = num**2

results = {}
t1 = threading.Thread(target=longSquare, args=(1, results))
t2 = threading.Thread(target=longSquare, args=(2, results))
t3 = threading.Thread(target=longSquare, args=(3, results))
t4 = threading.Thread(target=longSquare, args=(4, results))
t5 = threading.Thread(target=longSquare, args=(5, results))

t1.start()
t2.start()
t3.start()
t4.start()
t5.start()

t1.join()
t2.join()
t3.join()
t4.join()
t5.join()

print(results)
```

```python
def longSquare(num, results):
    time.sleep(1)
    results[num] = num**2

results = {}
threads = [threading.Thread(target=longSquare, args=(n, results)) for n in range(1, 4000)]
[t.start() for t in threads]
[t.join() for t in threads]
print((results))
```
